```python
def remove_alpha(cell):
    if isinstance(cell, str):
        return re.sub(r'[A-Za-z]', '', cell)
    return cell
```

```python
NewYork_100 = NewYork_100.applymap(remove_alpha)
NewYork_100 = NewYork_100.drop(0).reset_index(drop=True)

NewYork_100.tail()
```

```python
nan_count_per_column = NewYork_100.isna().sum()
print("NaN values per column:")
print(nan_count_per_column.iloc[20:])

# D2_MeanT_degC      1
# C3_MeanT_degC      1
# C3_MinT_degC       1
# C3_MaxT_degC       1
# J_D-IGH_ft        40
```

```python
# Mode for columns with less than 5 NaN values

for column in NewYork_100.columns:
    if NewYork_100[column].isna().sum() == 1:  
        mode_value = NewYork_100[column].mode().iloc[0]  
        NewYork_100[column] = NewYork_100[column].fillna(mode_value)
```

```python
NewYork_100['J_D-IGH_ft'] = pd.to_numeric(NewYork_100['J_D-IGH_ft'], errors='coerce')
J_mean = NewYork_100['J_D-IGH_ft'].mean()
NewYork_100['J_D-IGH_ft'] = NewYork_100['J_D-IGH_ft'].fillna(J_mean)

# Mean for D-IGH
```

```python
NewYork_100['C3_MaxT_degC'] = pd.to_numeric(NewYork_100['C3_MaxT_degC'], errors='coerce')
J_mean = NewYork_100['C3_MaxT_degC'].mean()
NewYork_100['C3_MaxT_degC'] = NewYork_100['C3_MaxT_degC'].fillna(J_mean)

NewYork_100['C3_MinT_degC'] = pd.to_numeric(NewYork_100['C3_MinT_degC'], errors='coerce')
J_mean = NewYork_100['C3_MinT_degC'].mean()
NewYork_100['C3_MinT_degC'] = NewYork_100['C3_MinT_degC'].fillna(J_mean)

NewYork_100['C3_MeanT_degC'] = pd.to_numeric(NewYork_100['C3_MeanT_degC'], errors='coerce')
J_mean = NewYork_100['C3_MeanT_degC'].mean()
NewYork_100['C3_MeanT_degC'] = NewYork_100['C3_MeanT_degC'].fillna(J_mean)

NewYork_100['D2_MaxT_degC'] = pd.to_numeric(NewYork_100['D2_MaxT_degC'], errors='coerce')
J_mean = NewYork_100['D2_MaxT_degC'].mean()
NewYork_100['D2_MaxT_degC'] = NewYork_100['D2_MaxT_degC'].fillna(J_mean)

NewYork_100['D2_MinT_degC'] = pd.to_numeric(NewYork_100['D2_MinT_degC'], errors='coerce')
J_mean = NewYork_100['D2_MinT_degC'].mean()
NewYork_100['D2_MinT_degC'] = NewYork_100['D2_MinT_degC'].fillna(J_mean)

NewYork_100['D2_MeanT_degC'] = pd.to_numeric(NewYork_100['D2_MeanT_degC'], errors='coerce')
J_mean = NewYork_100['D2_MeanT_degC'].mean()
NewYork_100['D2_MeanT_degC'] = NewYork_100['D2_MeanT_degC'].fillna(J_mean)
```

```python
NewYork_100 = NewYork_100.round(1)
# Remove '[4]' from all cells in the DataFrame
NewYork_100 = NewYork_100.applymap(lambda x: x.replace(' [4]', '') if isinstance(x, str) else x)
```
